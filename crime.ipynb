### **Why Crime Rates?**
#### Analyzing the crime rates of each neighborhood allows us to see which neighboorhood is safest. Safety is key to determining which neighboorhood is best to live in as it ensures security, security, and overall satisfaction from neighborhood to neighborhood. 

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
crime = pd.read_csv("crime_rates.csv")
neighborhood = crime['INCIDENTNEIGHBORHOOD'].value_counts()
crime_counts_df = neighborhood.reset_index()
crime_counts_df.columns = ['Neighborhood','Crime_Count']
crime_counts_df.tail(10)
```

```python
bottom_10_crime = crime_counts_df.tail(10)
bottom_10_crime.plot.bar(x='Neighborhood', y='Crime_Count', rot=65, title="Bottom 10 Crime Counts by Neighborhood")
```

```python
top_10_crime = crime_counts_df.head(10)
top_10_crime.plot.bar(x='Neighborhood', y='Crime_Count', rot=65, title="Top 10 Crime Counts by Neighborhood")
```
