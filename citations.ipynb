## Non-traffic Citations

#### Project Goal


I will analysis Non-traffic Citations in the neighbor hoods of Pittsburgh to as a submetric to conclude the best neighborhood in Pittsburgh. I will do this analyisis with a dataset from WPRDC. 

I have chosen non-traffic citations as a metric because it shows all the crimial offenses where law enforcement needed to issue a citation. I believe safety is the most important rule for a neighbor hood to be considered the best.

https://data.wprdc.org/dataset/non-traffic-citations/resource/6b11e87d-1216-463d-bbd3-37460e539d86

#### The Results

For this metric alone, the safest neighborhood is Outside County. I have found this result by counting every citation given per neighborhood. Outside County had the lowest number of citations given.

```python
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
df = pd.read_csv("nontraffic_citations.csv")
citation_counts = df['NEIGHBORHOOD'].value_counts()
citation_counts_df = citation_counts.reset_index()
citation_counts_df.columns = ['Neighborhood', 'Citation_Count']
print(citation_counts_df)
```

```python
bottom_10 = citation_counts_df.tail(10)
bottom_10.plot.bar(x='Neighborhood', y='Citation_Count', rot=65, title="Bottom 10 Citation Counts by Neighborhood")
```
